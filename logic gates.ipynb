PROBLEM STATEMENT: Implement simple logic network using MP neuron model

Logic gates are the basic building blocks of any digital system. It is an electronic circuit having one or more than one input and only one output. The relationship between the input and the output is based on a certain logic.

The early model of an artificial neuron is introduced by Warren McCulloch and Walter Pitts in 1943. The McCulloch-Pitts neural model is also known as linear threshold gate. It is a neuron of a set of inputs  $ X_1, X_2, X_3, ..., X_m $ and one output  $ y
$. The linear threshold gate simply classifies the set of inputs into two different classes. Thus the output  $ y
$ is binary. Such a function can be described mathematically using these equations:

 $\displaystyle Sum = \mathop{\sum_{i = 1}^{N}}{X_{i} W_{i}} + b, $

 $\displaystyle y = f(Sum). $

$ W_1, W_2, W_3, ..., W_m$ are weight values normalized in the range of either $ (0,1)$ or $ (-1,1)$ and associated with each input line,  $ Sum$ is the weighted sum, and  $ T$ is a threshold constant


The function  $ f$ is a linear step function at threshold  $ T$

```python
# function for and gate 
def and_gate(no1,no2):
    weight1 = 1
    weight2 = 1
    bias = -2
    ans1=gates(no1,no2,weight1,weight2,bias)
    return(ans1)
```

```python
# function for or gate 
def or_gate(no1,no2):
    weight1 = 1
    weight2 = 1
    bias = -1
    ans2=gates(no1,no2,weight1,weight2,bias)
    return(ans2)
```

```python
# function for not gate 
def not_gate(no1):
    no2 = 0 
    weight1 = -1
    weight2 = 0 
    bias = 0
    ans3 = gates(no1,no2,weight1,weight2,bias)
    return(ans3)
```

```python
# function for xor gate 
def xor_gate(no1,no2):
        a=and_gate(no1,no2)
        c=or_gate(no1,no2)
        b=not_gate(a)
        d=and_gate(b,c)
        return(d)
```

```python
# function for implementing MP neuron model 
def gates(no1,no2,weight1,weight2,bias):
    MP_neuron = weight1 * no1 + weight2 * no2 + bias
    output = int(MP_neuron >= 0)
    return(output)
```
